```python
## Crawling Data  for seen and unseen
i = 0
while(i <= 8000):
    suffix = str(code_list[i])
    suffix = suffix[2:-2]
    driver.get(new_url+suffix)
    try:
        Region = driver.find_element(By.ID, "RegionComponent")
        Region = Region.get_attribute('outerHTML')
        Region = get_item_between_strongs(Region)
        Region_list.append(Region)
    except NoSuchElementException:
        Region_list.append("Missing")
        
    try:
        Type = driver.find_element(By.ID, "TypeComponent")
        Type = Type.get_attribute('outerHTML')
        Type = get_item_between_strongs(Type)
        Type_list.append(Type)
    except NoSuchElementException:
        Type_list.append("Missing")
    
    try:
        Sqm = driver.find_element(By.ID, "Sq.MetersComponent")
        Sqm = Sqm.get_attribute('outerHTML')
        Sqm = get_item_between_strongs(Sqm)
        Sqm_list.append(Sqm)
    except NoSuchElementException:
        try:
            Sqm = driver.find_element(By.ID, "LandsizeComponent")
            Sqm = Sqm.get_attribute('outerHTML')
            Sqm = get_item_between_strongs(Sqm)
            Sqm_list.append(Sqm)
        except NoSuchElementException:
            Sqm_list.append("Missing")
            
    try:
        Year_of_const = driver.find_element(By.ID, "YearofconstructionComponent")
        Year_of_const = Year_of_const.get_attribute('outerHTML')
        Year_of_const = get_item_between_strongs(Year_of_const)
        Year_of_const_list.append(Year_of_const)
    except NoSuchElementException:
        Year_of_const_list.append("No Const Land")
        
    try:   
        Dist_from_sea = driver.find_element(By.ID, "DistancefromseaComponent")
        Dist_from_sea = Dist_from_sea.get_attribute('outerHTML')
        Dist_from_sea = get_item_between_strongs(Dist_from_sea)
        Dist_from_sea_list.append(Dist_from_sea)                                      
    except NoSuchElementException:
        Dist_from_sea_list.append("Missing")
    
    try:
        Dist_from_air = driver.find_element(By.ID, "DistancefromairportComponent")
        Dist_from_air = Dist_from_air.get_attribute('outerHTML')
        Dist_from_air = get_item_between_strongs(Dist_from_air)
        Dist_from_air_list.append(Dist_from_air)
    except NoSuchElementException:
        Dist_from_air_list.append("Missing")
        
    try: 
        price = driver.find_element(By.TAG_NAME, "h2")
        price = price.get_attribute('outerHTML')
        price = extract_number(price)
        price_list.append(price)
    except NoSuchElementException:
        price_list.append("Missing")
        
    i+=1

    
####################################UNSEEN DATA CRAWL###############################################################
i = 0
while(i < len(code_list_unseen)):
    suffix = str(code_list_unseen[i])
    driver.get(new_url+suffix)
    try:
        Region = driver.find_element(By.ID, "RegionComponent")
        Region = Region.get_attribute('outerHTML')
        Region = get_item_between_strongs(Region)
        Region_Unseen_list.append(Region)
    except NoSuchElementException:
        Region_Unseen_list.append("Missing")
        
    try:
        Type = driver.find_element(By.ID, "TypeComponent")
        Type = Type.get_attribute('outerHTML')
        Type = get_item_between_strongs(Type)
        Type_Unseen_list.append(Type)
    except NoSuchElementException:
        Type_Unseen_list.append("Missing")
    
    try:
        Sqm = driver.find_element(By.ID, "Sq.MetersComponent")
        Sqm = Sqm.get_attribute('outerHTML')
        Sqm = get_item_between_strongs(Sqm)
        Sqm_Unseen_list.append(Sqm)
    except NoSuchElementException:
        try:
            Sqm = driver.find_element(By.ID, "LandsizeComponent")
            Sqm = Sqm.get_attribute('outerHTML')
            Sqm = get_item_between_strongs(Sqm)
            Sqm_Unseen_list.append(Sqm)
        except NoSuchElementException:
            Sqm_Unseen_list.append("Missing")
            
    try:
        Year_of_const = driver.find_element(By.ID, "YearofconstructionComponent")
        Year_of_const = Year_of_const.get_attribute('outerHTML')
        Year_of_const = get_item_between_strongs(Year_of_const)
        Year_of_const_Unseen_list.append(Year_of_const)
    except NoSuchElementException:
        Year_of_const_Unseen_list.append("No Const Land")
        
    try:   
        Dist_from_sea = driver.find_element(By.ID, "DistancefromseaComponent")
        Dist_from_sea = Dist_from_sea.get_attribute('outerHTML')
        Dist_from_sea = get_item_between_strongs(Dist_from_sea)
        Dist_from_sea_Unseen_list.append(Dist_from_sea)                                      
    except NoSuchElementException:
        Dist_from_sea_Unseen_list.append("Missing")
    
    try:
        Dist_from_air = driver.find_element(By.ID, "DistancefromairportComponent")
        Dist_from_air = Dist_from_air.get_attribute('outerHTML')
        Dist_from_air = get_item_between_strongs(Dist_from_air)
        Dist_from_air_Unseen_list.append(Dist_from_air)
    except NoSuchElementException:
        Dist_from_air_Unseen_list.append("Missing")
        
    try: 
        price = driver.find_element(By.TAG_NAME, "h2")
        price = price.get_attribute('outerHTML')
        price = extract_number(price)
        price_Unseen_list.append(price)
    except NoSuchElementException:
        price_Unseen_list.append("Missing")
        
    i+=1
```
